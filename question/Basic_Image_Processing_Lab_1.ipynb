# 1. Import Required Libraries


````python

import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

print("Libraries imported successfully!")
````


# 2. Reading an Image Using OpenCV

> Replace `'sample.jpg'` with your own image file.


````python

image_cv = cv2.imread('sample.jpg')

print("Image loaded successfully!")
print("Image Shape:", image_cv.shape)
````


# 3. Displaying an Image

OpenCV reads images in **BGR** format, while Matplotlib expects **RGB**.


````python

image_rgb = cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Image using OpenCV + Matplotlib")
plt.axis('off')
plt.show()
````


# 4. Reading an Image Using Pillow (PIL)


````python

image_pil = Image.open('sample.jpg')

print("Image Size:", image_pil.size)
print("Image Mode:", image_pil.mode)

image_pil
````


# 5. Understanding Pixels

A digital image is made up of tiny elements called **pixels**.

For RGB images:
- Red channel
- Green channel
- Blue channel


````python

# Accessing a pixel value
pixel = image_rgb[100, 100]

print("Pixel value at (100,100):", pixel)
````


# 6. Exploring Image Properties


````python

height, width, channels = image_rgb.shape

print("Height:", height)
print("Width:", width)
print("Channels:", channels)
````


# 7. Splitting Color Channels


````python

r = image_rgb[:, :, 0]
g = image_rgb[:, :, 1]
b = image_rgb[:, :, 2]

fig, ax = plt.subplots(1, 3, figsize=(12,4))

ax[0].imshow(r, cmap='Reds')
ax[0].set_title("Red Channel")

ax[1].imshow(g, cmap='Greens')
ax[1].set_title("Green Channel")

ax[2].imshow(b, cmap='Blues')
ax[2].set_title("Blue Channel")

for a in ax:
    a.axis('off')

plt.show()
````


# 8. Converting RGB Image to Grayscale


````python

gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
````


# 9. Image Resizing


````python

resized = cv2.resize(image_rgb, (300, 300))

plt.imshow(resized)
plt.title("Resized Image")
plt.axis('off')
plt.show()
````


# 10. Image Cropping


````python

cropped = image_rgb[50:250, 50:250]

plt.imshow(cropped)
plt.title("Cropped Image")
plt.axis('off')
plt.show()
````


# 11. Image Quantization

Quantization reduces the number of intensity levels in an image.

Normally grayscale images have:

256 intensity levels

because 8-bit grayscale uses values from 0–255.

Quantization reduces these levels.

This is useful in:
- Image compression
- Low-memory devices
- Artistic effects


````python

# Reduce grayscale levels

quantized = gray // 64 * 64

plt.imshow(quantized, cmap='gray')
plt.title("Quantized Image")
plt.axis('off')
plt.show()
````


# 12. Simple Image Manipulation

## Inverting an Image


````python

inverted = 255 - image_rgb

plt.imshow(inverted)
plt.title("Inverted Image")
plt.axis('off')
plt.show()
````


# 13. Saving an Image


````python

cv2.imwrite('output_image.jpg', cv2.cvtColor(resized, cv2.COLOR_RGB2BGR))

print("Image saved successfully!")
````
