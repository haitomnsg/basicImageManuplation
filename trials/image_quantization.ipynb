```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
gray = cv2.imread('cat_gray.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(gray, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
quantized = gray // 64 * 64

plt.imshow(quantized, cmap='gray')
plt.title('Quantized Image')
plt.axis('off')
plt.show()
```
