# Import OpenCV

```python
import cv2
import matplotlib.pyplot as plt
```

# Read Image

```python
img = cv2.imread('cat_small.png')
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

new_width = 512
new_height = 512
```

# Image Resizing

```python
nearest = cv2.resize(image_rgb, (new_width, new_height), interpolation=cv2.INTER_NEAREST)

linear = cv2.resize(image_rgb, (new_width, new_height), interpolation=cv2.INTER_LINEAR)

cubic = cv2.resize(image_rgb, (new_width, new_height), interpolation=cv2.INTER_CUBIC)

area = cv2.resize(image_rgb, (new_width, new_height), interpolation=cv2.INTER_AREA)

lanczos = cv2.resize(image_rgb, (new_width, new_height), interpolation=cv2.INTER_LANCZOS4)
```

# Displaying Commands

```python
titles = [
    "Original Image",
    "INTER_NEAREST",
    "INTER_LINEAR",
    "INTER_CUBIC",
    "INTER_AREA",
    "INTER_LANCZOS4"
]

images = [
    image_rgb,
    nearest,
    linear,
    cubic,
    area,
    lanczos
]

plt.figure(figsize=(15, 10))

for i in range(len(images)):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

# Cropping Image

```python
y_start = 0
y_end = 30
x_start = 0
x_end = 30

cropped_image = image_rgb[y_start:y_end, x_start:x_end]

plt.figure(figsize=(6, 6))
plt.imshow(cropped_image)
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```
